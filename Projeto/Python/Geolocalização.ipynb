```python
from pycep_correios import get_address_from_cep, WebService
from geopy.geocoders import Nominatim
from geopy.extra.rate_limiter import RateLimiter
import pandas as pd

geolocator = Nominatim(user_agent="updt_geolocalizacao")
geocode = RateLimiter(geolocator.geocode, min_delay_seconds=1)

path_file = '/content/drive/MyDrive/Dataset/olist_geolocation_dataset.csv'
df = pd.read_csv(path_file, sep=',')

list_cep = df['geolocation_zip_code_prefix'].apply(lambda cep: f'0{cep}000').unique()
list_update = []

for cep in list_cep:
  try:

    address = get_address_from_cep(cep, webservice=WebService.CORREIOS)
    location = geolocator.geocode(address.get('logradouro'))
    list_update.append([cep, location.latitude, location.longitude])

  except BaseException as exc:
    continue

df = pd.DataFrame(list_update, columns=['cep', 'up_lat', 'up_long'])
df.to_csv('/content/drive/MyDrive/Dataset/updt_geolocation.csv')
```
