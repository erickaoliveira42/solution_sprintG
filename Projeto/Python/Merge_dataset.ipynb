```python
import pandas as pd
```

```python
path_file = '/content/drive/MyDrive/Dataset/olist_geolocation_dataset.csv'
df_olist = pd.read_csv(path_file, sep=',')

path_file = '/content/drive/MyDrive/Dataset/updt_geolocation.csv'
df_update = pd.read_csv(path_file, sep=',')
```

```python
df_olist['geolocation_zip_code_prefix'] = df_olist['geolocation_zip_code_prefix'].apply(pd.to_numeric) 
df_olist
```

```python
df_update.rename(columns={'cep': 'geolocation_zip_code_prefix'}, inplace = True)
if len(df_update['geolocation_zip_code_prefix']) == 7:
  df_update['geolocation_zip_code_prefix'] = df_update['geolocation_zip_code_prefix'].apply(str).str.slice( 0, 3)
else:
  df_update['geolocation_zip_code_prefix'] = df_update['geolocation_zip_code_prefix'].apply(str).str.slice( 0, 4)
df_update['geolocation_zip_code_prefix'] = df_update['geolocation_zip_code_prefix'].apply(pd.to_numeric) 
df_update
```

```python
print(df_olist.dtypes)
print(df_update.dtypes)
```

```python
df_merge = pd.merge(df_olist, df_update, on = 'geolocation_zip_code_prefix', how='left')
df_merge['up_lat'].fillna(0, inplace=True)
df_merge['up_long'].fillna(0, inplace=True)

df_merge
```

```python
df_extracao = pd.DataFrame()

df_extracao['geolocation_zip_code_prefix'] = df_merge['geolocation_zip_code_prefix']
df_extracao['geolocation_lat'] = df_merge['up_lat']
df_extracao['geolocation_lng'] = df_merge['up_long']
df_extracao['geolocation_city'] = df_merge['geolocation_city']
df_extracao['geolocation_state'] = df_merge['geolocation_state']
df_extracao['geolocation_lat_old'] = df_merge['geolocation_lat']
df_extracao['geolocation_lng_old'] = df_merge['geolocation_lng']

df_extracao.to_csv('/content/drive/MyDrive/Dataset/olist_geolocation_dataset_updt.csv')
df_extracao
```
